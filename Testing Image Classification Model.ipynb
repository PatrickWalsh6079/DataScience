```python
# maximum image size: 10 MB
# maximum image count: 20
# for zip files: max 100 MB

import json, xlwt, os
from watson_developer_cloud import VisualRecognitionV3
from xlwt import Workbook
from xlwt import *
from datetime import date

###############################################
# invoke model with version number and API key
##############################################
visual_recognition = VisualRecognitionV3(
    'INSERT MODEL DATE HERE',
    iam_apikey='INSERT YOUR API KEY HERE')


# counter used when writing values to spreadsheet
excel_counter = 0


#################################################################################################
# specify parameters for iteration number, Testing images, desired threshold, and results export
iteration = 1
test_folder = r'C:\Users\pwalsh\Desktop\PROJECTS\Visual Recognition\Helena\Testing'
threshold = 0.80
notebook_results = r'C:\Users\pwalsh\Desktop\PROJECTS\Visual Recognition\Helena\Results\Notebook Tomato 2 Results 1.xls'
class1 = 'HEALTHY'
class2 = 'MITES'
class3 = 'MOLD'
#################################################################################################



######################################################################
# create excel spreadsheet and add worksheets 'Results' and 'Analysis'
######################################################################
wb = Workbook()
results = wb.add_sheet('Results')

# write column headers for 'Results' sheet
results.write(0, 0, 'Date') 
results.write(0, 1, 'Model Name') 
results.write(0, 2, 'Iteration') 
results.write(0, 3, 'File Name') 
results.write(0, 4, 'Type') 
results.write(0, 5, class1) 
results.write(0, 6, class2)
results.write(0, 7, class3)
results.write(0, 8, 'Prediction')
results.write(0, 10, 'Threshold')
results.write(0, 12, '11')
results.write(0, 13, '21')
results.write(0, 14, '31')
results.write(0, 15, 'U1')
results.write(0, 17, '12')
results.write(0, 18, '22')
results.write(0, 19, '32')
results.write(0, 20, 'U2')
results.write(0, 22, '13')
results.write(0, 23, '23')
results.write(0, 24, '33')
results.write(0, 25, 'U3')




############################################################
# grab zip folder of test images and initiate tagging event
############################################################
for f in os.listdir(test_folder):
    f = os.path.join(test_folder, f)
    
    with open(f, 'rb') as images_file:
        classes = visual_recognition.classify(images_file,threshold='0.0',owners=["me"]).get_result()
        #print(json.dumps(classes, indent=2))
    
    
    
    ########################################
    # these values will be written to Excel
    #######################################
    model_name = json.dumps(classes["images"][0]["classifiers"][0]["name"])
    file_name = json.dumps(classes["images"][0]["image"])
    c1_score = float(json.dumps(classes["images"][0]["classifiers"][0]["classes"][0]["score"]))
    c2_score = float(json.dumps(classes["images"][0]["classifiers"][0]["classes"][1]["score"]))
    c3_score = float(json.dumps(classes["images"][0]["classifiers"][0]["classes"][2]["score"]))
    test_date = str(date.today())
    print('Model: '+model_name)
    print('Image: '+file_name)
    print(class1+': ',c1_score)
    print(class2+': ',c2_score)
    print(class3+': ',c3_score)
        
        
        
        
    # determine type of image
    image_type = ''
    if  class1 in file_name:
        print('Type: '+class1)
        image_type = class1
    elif class2 in file_name:
        print('Type: '+class2)
        image_type = class2
    elif class3 in file_name:
        print('Type: '+class3)
        image_type = class3
        
        
            
    excel_counter += 1
        
        
        
    # write results to Excel spreadsheet
    results.write(excel_counter, 0, test_date) 
    results.write(excel_counter, 1, model_name) 
    results.write(excel_counter, 2, iteration) 
    results.write(excel_counter, 3, file_name) 
    results.write(excel_counter, 4, image_type) 
    results.write(excel_counter, 5, c1_score) 
    results.write(excel_counter, 6, c2_score)
    results.write(excel_counter, 7, c3_score)
    # prediction
    results.write(excel_counter, 8, Formula('IF(F{}>=K2,"{}",IF(G{}>=K2,"{}",IF(H{}>=K2,"{}","none")))'
                                            .format(excel_counter+1,class1,
                                                    excel_counter+1,class2,
                                                    excel_counter+1,class3,)))
    # cells that tally the 11,21,31,U1, etc
    results.write(excel_counter, 12, Formula('IF(I{}="{}";IF(E{}="{}";1;0);0)'
                                             .format(excel_counter+1,class1,
                                                    excel_counter+1,class1))) 
    results.write(excel_counter, 13, Formula('IF(I{}="{}";IF(E{}="{}";1;0);0)'
                                             .format(excel_counter+1,class2,
                                                    excel_counter+1,class1))) 
    results.write(excel_counter, 14, Formula('IF(I{}="{}";IF(E{}="{}";1;0);0)'
                                             .format(excel_counter+1,class3,
                                                     excel_counter+1,class1))) 
    results.write(excel_counter, 15, Formula('IF(I{}="none";IF(E{}="{}";1;0);0)'
                                             .format(excel_counter+1,excel_counter+1,class1)))
    results.write(excel_counter, 17, Formula('IF(I{}="{}";IF(E{}="{}";1;0);0)'
                                             .format(excel_counter+1,class1,
                                                     excel_counter+1,class2)))
    results.write(excel_counter, 18, Formula('IF(I{}="{}";IF(E{}="{}";1;0);0)'
                                             .format(excel_counter+1,class2,
                                                     excel_counter+1,class2)))
    results.write(excel_counter, 19, Formula('IF(I{}="{}";IF(E{}="{}";1;0);0)'
                                             .format(excel_counter+1,class3,
                                                     excel_counter+1,class2)))
    results.write(excel_counter, 20, Formula('IF(I{}="none";IF(E{}="{}";1;0);0)'
                                             .format(excel_counter+1,excel_counter+1,class2)))
    results.write(excel_counter, 22, Formula('IF(I{}="{}";IF(E{}="{}";1;0);0)'
                                             .format(excel_counter+1,class1,
                                                     excel_counter+1,class3)))                                       
    results.write(excel_counter, 23, Formula('IF(I{}="{}";IF(E{}="{}";1;0);0)'
                                             .format(excel_counter+1,class2,
                                                     excel_counter+1,class3)))                                       
    results.write(excel_counter, 24, Formula('IF(I{}="{}";IF(E{}="{}";1;0);0)'
                                             .format(excel_counter+1,class3,
                                                     excel_counter+1,class3)))                                       
    results.write(excel_counter, 25, Formula('IF(I{}="none";IF(E{}="{}";1;0);0)'
                                             .format(excel_counter+1,excel_counter+1,class3)))                                       
    
    
    
    
    print('\n')
        


# create metrics worksheet
metrics = wb.add_sheet('Metrics')

# create confusion matrix
metrics.write(0, 2, 'Confusion Matrix')
metrics.write(2, 2, 'Actual')
metrics.write(3, 2, class1)
metrics.write(3, 3, class2)
metrics.write(3, 4, class3)
metrics.write(4, 0, 'Predicted')
metrics.write(4, 1, class1)
metrics.write(5, 1, class2)
metrics.write(6, 1, class3)
metrics.write(7, 1, 'UNCLASSIFIED')
metrics.write(4, 2, Formula('SUM(Results!M2:M{})'.format(excel_counter+1)))
metrics.write(4, 3, Formula('SUM(Results!R2:R{})'.format(excel_counter+1)))
metrics.write(4, 4, Formula('SUM(Results!W2:W{})'.format(excel_counter+1)))
metrics.write(5, 2, Formula('SUM(Results!N2:N{})'.format(excel_counter+1)))
metrics.write(5, 3, Formula('SUM(Results!S2:S{})'.format(excel_counter+1)))
metrics.write(5, 4, Formula('SUM(Results!X2:X{})'.format(excel_counter+1)))
metrics.write(6, 2, Formula('SUM(Results!O2:O{})'.format(excel_counter+1)))
metrics.write(6, 3, Formula('SUM(Results!T2:T{})'.format(excel_counter+1)))
metrics.write(6, 4, Formula('SUM(Results!Y2:Y{})'.format(excel_counter+1)))
metrics.write(7, 2, Formula('SUM(Results!P2:P{})'.format(excel_counter+1)))
metrics.write(7, 3, Formula('SUM(Results!U2:U{})'.format(excel_counter+1)))
metrics.write(7, 4, Formula('SUM(Results!Z2:Z{})'.format(excel_counter+1)))



# calculate precision, recall, F1, and accuracy
metrics.write(9, 2, 'Precision')
metrics.write(9, 3, 'Recall')
metrics.write(9, 4, 'F1')
metrics.write(10, 1, class1)
metrics.write(11, 1, class2)
metrics.write(12, 1, class3)
metrics.write(3, 6, 'Accuracy')
metrics.write(3, 7, Formula('(C5+D6+E7)/(C5+D5+E5+C6+D6+E6+C7+D7+E7+C8+D8+E8)'))
metrics.write(10, 2, Formula('C5/(C5+D5+E5)'))
metrics.write(10, 3, Formula('C5/(C5+C6+C7+C8)'))
metrics.write(10, 4, Formula('2*(C11*D11)/(C11+D11)'))
metrics.write(11, 2, Formula('D6/(C6+D6+E6)'))
metrics.write(11, 3, Formula('D6/(D5+D6+D7+D8)'))
metrics.write(11, 4, Formula('2*(C12*D12)/(C12+D12)'))
metrics.write(12, 2, Formula('E7/(C7+D7+E7)'))
metrics.write(12, 3, Formula('E7/(E5+E6+E7+E8)'))
metrics.write(12, 4, Formula('2*(C13*D13)/(C13+D13)'))




# create analysis worksheet
analysis = wb.add_sheet('Analysis')


# class1 results
analysis.write(0, 1, '{} Classification results'.format(class1))
analysis.write(1, 0, 'Test')
analysis.write(1, 1, 'Actual positives')
analysis.write(1, 2, 'Classified positive (correctly)')
analysis.write(1, 3, 'Actual negatives')
analysis.write(1, 4, 'Classified negative (correctly)')
analysis.write(1, 7, 'Probability of correct classification')
analysis.write(1, 8, 'Probability of False Positive classification')
analysis.write(1, 9, 'Probability of False Negative classification')
analysis.write(1, 12, 'Threshold')
analysis.write(2, 0, iteration)
analysis.write(2, 1, Formula('Metrics!C5+Metrics!C6+Metrics!C7+Metrics!C8'))
analysis.write(2, 2, Formula('Metrics!C5'))
analysis.write(2, 3, Formula('Metrics!D5+Metrics!D6+Metrics!D7+Metrics!D8+Metrics!E5+Metrics!E6+Metrics!E7+Metrics!E8'))
analysis.write(2, 4, Formula('Metrics!D6+Metrics!E6+Metrics!D7+Metrics!E7+Metrics!D8+Metrics!E8'))
analysis.write(2, 7, Formula('(C3+E3)/(B3+D3)'))
analysis.write(2, 8, Formula('(D3-E3)/D3'))
analysis.write(2, 9, Formula('(B3-C3)/B3'))
analysis.write(2, 12, threshold)
# class 2 results
analysis.write(11, 1, '{} Classification results'.format(class2))
analysis.write(12, 0, 'Test')
analysis.write(12, 1, 'Actual positives')
analysis.write(12, 2, 'Classified positive (correctly)')
analysis.write(12, 3, 'Actual negatives')
analysis.write(12, 4, 'Classified negative (correctly)')
analysis.write(12, 7, 'Probability of correct classification')
analysis.write(12, 8, 'Probability of False Positive classification')
analysis.write(12, 9, 'Probability of False Negative classification')
analysis.write(13, 0, iteration)
analysis.write(13, 1, Formula('Metrics!D5+Metrics!D6+Metrics!D7+Metrics!D8'))
analysis.write(13, 2, Formula('Metrics!D6'))
analysis.write(13, 3, Formula('Metrics!C5+Metrics!C6+Metrics!C7+Metrics!C8+Metrics!E5+Metrics!E6+Metrics!E7+Metrics!E8'))
analysis.write(13, 4, Formula('Metrics!C5+Metrics!E5+Metrics!C7+Metrics!E7+Metrics!C8+Metrics!E8'))
analysis.write(13, 7, Formula('(C14+E14)/(B14+D14)'))
analysis.write(13, 8, Formula('(D14-E14)/D14'))
analysis.write(13, 9, Formula('(B14-C14)/B14'))
# class 3 results
analysis.write(22, 1, '{} Classification results'.format(class3))
analysis.write(23, 0, 'Test')
analysis.write(23, 1, 'Actual positives')
analysis.write(23, 2, 'Classified positive (correctly)')
analysis.write(23, 3, 'Actual negatives')
analysis.write(23, 4, 'Classified negative (correctly)')
analysis.write(23, 7, 'Probability of correct classification')
analysis.write(23, 8, 'Probability of False Positive classification')
analysis.write(23, 9, 'Probability of False Negative classification')
analysis.write(24, 0, iteration)
analysis.write(24, 1, Formula('Metrics!E5+Metrics!E6+Metrics!E7+Metrics!E8'))
analysis.write(24, 2, Formula('Metrics!E7'))
analysis.write(24, 3, Formula('Metrics!C5+Metrics!C6+Metrics!C7+Metrics!C8+Metrics!D5+Metrics!D6+Metrics!D7+Metrics!D8'))
analysis.write(24, 4, Formula('Metrics!C5+Metrics!D5+Metrics!C6+Metrics!D6+Metrics!C8+Metrics!D8'))
analysis.write(24, 7, Formula('(C25+E25)/(B25+D25)'))
analysis.write(24, 8, Formula('(D25-E25)/D25'))
analysis.write(24, 9, Formula('(B25-C25)/B25'))
# connects threshold parameter in Results sheet to one in Analysis sheet
results.write(1, 10, Formula('Analysis!M3'))





# save spreadsheet
wb.save(notebook_results)
```
