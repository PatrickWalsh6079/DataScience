# Convolutional Neural Network

This simple CNN can recognize four different letters, 'X', 'O', '\', and '/' as
depicted in a simple 3x3 pixel image. It takes input in the form of one of the
above letters and generates an output in the form of a prediction for which 
letter it is through its Convolution, Pooling, and Fully-connected layers.

```python
grid_input = input("Would you like to display 'X', 'O', '\', or '/'?")
layout = []
print(layout)

if grid_input == 'x'.lower():
    print('X')
    layout = [1,-1,1,-1,1,-1,1,-1,1]
    print(layout)
elif grid_input == 'o'.lower():
    print('O')
    layout = [-1,1,-1,1,-1,1,-1,1,-1]
    print(layout)
elif grid_input == '\\'.lower():
    print('\\')
    layout = [1,-1,-1,-1,1,-1,-1,-1,1]
    print(layout)
elif grid_input == '/'.lower():
    print('/')
    layout = [-1,-1,1,-1,1,-1,1,-1,-1]
    print(layout)
else:
    pass
```

```python
threshold = 3

def convo_layer():
    
    # convolution layer
    filter_backslash = []
    convo_0 = +layout[0]-layout[1]-layout[3]+layout[4]
    #print(convo_0)
    filter_backslash.append(convo_0)
    convo_1 = +layout[1]-layout[2]-layout[4]+layout[5]
    #print(convo_1)
    filter_backslash.append(convo_1)
    convo_2 = +layout[3]-layout[4]-layout[6]+layout[7]
    #print(convo_2)
    filter_backslash.append(convo_2)
    convo_3 = +layout[4]-layout[5]-layout[7]+layout[8]
    #print(convo_3)
    filter_backslash.append(convo_3)
    #print('filter_blackslash: ',filter_backslash)
    
    filter_forwardslash = []
    convo_0 = -layout[0]+layout[1]+layout[3]-layout[4]
    #print(convo_0)
    filter_forwardslash.append(convo_0)
    convo_1 = -layout[1]+layout[2]+layout[4]-layout[5]
    #print(convo_1)
    filter_forwardslash.append(convo_1)
    convo_2 = -layout[3]+layout[4]+layout[6]-layout[7]
    #print(convo_2)
    filter_forwardslash.append(convo_2)
    convo_3 = -layout[4]+layout[5]+layout[7]-layout[8]
    #print(convo_3)
    filter_forwardslash.append(convo_3)
    #print('filter_forwardlash: ',filter_forwardslash)
    
    # pooling layer
    pooled_blackslash = []
    for item in filter_backslash:
        if item > threshold:
            pooled_blackslash.append('\\')
        else:
            pooled_blackslash.append(' ')
    
    #print('pooled_blackslash: ',pooled_blackslash)
    
    pooled_forwardslash = []
    for item in filter_forwardslash:
        if item > threshold:
            pooled_forwardslash.append('/')
        else:
            pooled_forwardslash.append(' ')
    
    #print('pooled_forwardslash: ',pooled_forwardslash)
    
    # fully connected layer
    fully_connected = []
    counter = 0
    
    # superimpose pooling layer outputs for fully connected layer
    for item in pooled_blackslash:
        fully_connected.append(item)
        
    for item in pooled_forwardslash:
        if item == ' ':
            counter += 1
        else:
            fully_connected[counter] = item
            counter += 1
    
    #print('fully_connected: ',fully_connected)
    
    back_slash = []
    forward_slash = []
    for item in fully_connected:
        if item == '\\':
            back_slash.append(1)
        else:
            back_slash.append(-1)
            
    #print('back_slash: ',back_slash)
    
    for item in fully_connected:
        if item == '/':
            forward_slash.append(1)
        else:
            forward_slash.append(-1)
            
    #print('forward_slash: ',forward_slash)
    
    # letter filters
    
    filter_x = +back_slash[0]-back_slash[1]-back_slash[2]+back_slash[3]-forward_slash[0]+forward_slash[1]+forward_slash[2]-forward_slash[3]
    filter_o = -back_slash[0]+back_slash[1]+back_slash[2]-back_slash[3]+forward_slash[0]-forward_slash[1]-forward_slash[2]+forward_slash[3]
    filter_fslash = -back_slash[0]-back_slash[1]-back_slash[2]-back_slash[3]-forward_slash[0]+forward_slash[1]+forward_slash[2]-forward_slash[3]
    filter_bslash = +back_slash[0]-back_slash[1]-back_slash[2]+back_slash[3]-forward_slash[0]-forward_slash[1]-forward_slash[2]-forward_slash[3]
    #print('filter_x: ',filter_x)
    #print('filter_o: ',filter_o)
    #print('filter_fslash: ',filter_fslash)
    #print('filter_bslash: ',filter_bslash)
    
    #if filter_x > filter_o and 
    prediction = (filter_x,filter_o,filter_fslash,filter_bslash)
    print('Prediction: ',prediction)
    
    if prediction[0] == max(prediction):
        print('This is an X')
    elif prediction[1] == max(prediction):
        print('This is an O')
    elif prediction[2] == max(prediction):
        print('This is a /')
    elif prediction[3] == max(prediction):
        print('This is a \\')

convo_layer()
```
